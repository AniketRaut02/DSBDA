```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
# Sample Sentences
sentence1 = "I will walk 500 miles and I would walk 500 more. Just to be the man who walked a thousand miles to fall down at your door!"

sentence2 = "I played the play playfully as the players were playing in the ground."
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize

print("---------Tokenized Words------------")
print('Tokenized words:', word_tokenize(sentence1))
print("\n")
```

```python
print("---------Tokenized Sentences------------")
print('Tokenized sentences:', sent_tokenize(sentence1))
print("\n")
```

```python
from nltk import pos_tag
print("---------POS Tagging------------")
token = word_tokenize(sentence1) + word_tokenize(sentence2)
print('POS tagged:', pos_tag(token))
print("\n")
```

```python
print("---------Stop-Words Removal------------")
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) # set is used to add the word into set for fast searching
token = word_tokenize(sentence1)
cleaned_token = []
for word in token:
   if word not in stop_words:
      cleaned_token.append(word)
print("Unclean version:", token)
print("\n")
print("Cleaned version:", cleaned_token)
print("\n")
```

```python
print("---------Stemming------------")
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()  # object to perform stemming
token = word_tokenize(sentence2)
stemmed = [stemmer.stem(word) for word in token]
print("Stemmed words:", stemmed)
print("\n")
```

```python
print("---------Lemmatization------------")
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()  #object to perform lemmatization
token = word_tokenize(sentence2)
lemmatized = [lemmatizer.lemmatize(word) for word in token]
print("Lemmatized words:", lemmatized)
print("\n")
```
